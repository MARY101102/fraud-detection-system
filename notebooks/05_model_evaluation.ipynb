## Evaluation Notes

The models were evaluated using fraud-focused metrics instead of accuracy.

### Important Metrics

- Precision
- Recall
- F1-score
- ROC-AUC
- PR-AUC
- Confusion matrix

### Why PR-AUC?

The dataset is highly imbalanced, so PR-AUC is more meaningful than accuracy.

### Why Recall?

Recall shows how many actual fraud transactions were correctly detected.

### Why Precision?

Precision shows how many predicted fraud transactions were actually fraud.

### Important Limitation

This evaluation uses the default classification threshold of `0.5`.

The next step is threshold tuning to find a better decision threshold for fraud detection.
