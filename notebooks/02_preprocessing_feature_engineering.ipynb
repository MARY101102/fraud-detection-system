## Feature Engineering Summary

In this step, new transaction-level features were created to improve model learning.

### Amount Features

- `amount_log`
- `amount_is_zero`
- `amount_is_high`

### Time Features

- `hour`
- `day`
- `day_of_week`
- `is_night`
- `is_weekend_simulated`

### Transaction Type Features

- `is_cash_out`
- `is_transfer`
- `is_payment`
- `is_debit`
- `is_cash_in`

### Balance Features

- `origin_balance_diff`
- `destination_balance_diff`
- `origin_balance_error`
- `destination_balance_error`
- `origin_balance_ratio`
- `destination_balance_ratio`
- `origin_old_balance_is_zero`
- `destination_old_balance_is_zero`

### Important Note

Some balance-based features may use post-transaction information.  
These features are useful for experimentation, but in a real-time fraud detection system they should be reviewed carefully to avoid data leakage.
