## Model Training Notes

The trained models were evaluated using fraud-focused metrics instead of accuracy.

Important metrics:

- Precision
- Recall
- F1-score
- ROC-AUC
- PR-AUC
- Confusion matrix values

### Why PR-AUC is important

Fraud datasets are highly imbalanced. PR-AUC is more useful than accuracy because it focuses on how well the model finds the minority fraud class.

### Why recall is important

Recall shows how many real fraud transactions were correctly detected.

### Why precision is important

Precision shows how many predicted fraud transactions were actually fraud.

### Important Rule

The test set was not resampled.  
Only training data was sampled or class-weighted.
