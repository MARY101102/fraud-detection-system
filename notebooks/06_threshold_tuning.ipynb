## Threshold Tuning Notes

The default threshold of `0.5` may not be optimal for fraud detection.

### Lower Threshold

A lower threshold usually catches more fraud cases.

Benefit:

- Higher recall

Risk:

- More false positives

### Higher Threshold

A higher threshold usually reduces false alarms.

Benefit:

- Higher precision

Risk:

- More false negatives

### Selected Threshold

The selected threshold in this experiment is based on the highest F1-score.

In a real banking system, the final threshold should be selected based on:

- Cost of missed fraud
- Cost of false alarms
- Analyst review capacity
- Customer experience
